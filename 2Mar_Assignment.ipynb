```python
## Matplotlib is a data visualization library in Python that allows users to create various types of static, interactive, and animated plots. 
## It provides a wide range of tools for creating 2D and 3D plots,including line plots, scatter plots, bar plots, histogram plots, pie charts, and many more.
```

```python
## 1. Line Plots: It is used to display trends or changes in data.
## 2. Scatter Plots: It is used to display the relationship between variables
## 3. Bar Plots: It is used to display categorical data by plotting bars with heights or lengths
## 4. Histogram Plot: It is used to display the distribution of a single variable by grouping it into bins.
## 5. Pie Charts: It is used to display the compostion of a categorical variable.
```

```python
## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
## Note: Also add title, xlabel, and ylabel to the plot.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("This is my Scatter Plot using Matplotlib ")
plt.xlabel("This is my x-label")
plt.ylabel("This is my y-label")
plt.show()

```

```python
## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

## Subplot() Function:: This function is used to create multiple plots in a single frame.It takes 3 arguments: 
                 ## Number of rows, Number of columns and Index of the plot being created.
import numpy as np
import matplotlib.pyplot as plt


x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("Line 1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("Line 2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("Line 3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("Line 4")

plt.suptitle("Four Line Plots")
plt.show()
```

```python
## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(company,profit)
plt.show()
```

```python
## A bar plot is a type of graph that displays data using rectangular bars with lengths proportional to the values they represent.
## A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.
```

```python
## Matplotlib API provides the bar() function that can be used in the MATLAB style use as well as object oriented API.
```

```python
## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
import numpy as np
import matplotlib.pyplot as plt

# generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# create a list of the two datasets
data = [box1, box2]

# plot the box plot
plt.boxplot(data)

# set the labels for the x-axis
plt.xticks([1, 2], ['Box 1', 'Box 2'])


# add a title to the plot
plt.title('Box Plot of Two Datasets')

# show the plot
plt.show()








```
