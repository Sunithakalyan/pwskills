```python
 ## Seaborn is a visualization Library in python high level interfaces for creating informative and attractive statistical graphs.
 ## There are 5 plots in Seaborn. 1. Scatter plot, 2.Line Plot, 3. Histogram, 4. Box Plot, 5. Heatmap
 ## 1. Scatter Plot: It is used to display the relationship between 2 variables.
 ## 2. Line Plot:  Series if data points connected by straight lines.
 ## 3. Histogram: Graphical representation of the distribution of the data.
 ## 4. Box Plot: It is used to display the distribution of a dataset. It shows the median, quartiles, and outliers of the data.  
 ## 5. Heatmap: Graphical representation of the data where the values are represented as colors.

 
```

```python
## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" 
 ## for different events and regions. Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri .head()
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal)
```

```python
## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
## Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.age )
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.fare)
```

```python
## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. 
## Use the hue parameter for the 'cut' column of the diamonds dataset.

diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python

diamonds.hist(column='price', bins=50)
```

```python
ax = sns.barplot(x = 'cut', y = 'price' , hue = 'color' , data=diamonds )
```
